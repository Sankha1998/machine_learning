```python
x=data['YearsExperience'].values
y=data['Salary'].values
plt.scatter(x, y)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
x_mean=np.mean(x)
y_mean=np.mean(y)
```

```python
## calculation of line's m and b 

x_num = x-x_mean
y_num = y-y_mean
m=(sum(x_num*y_num))/(sum(x_num*x_num))
```

```python
m
```

```python
b=y_mean-(m*x_mean)
```

```python
b
```

```python
y1 = m*x+b
plt.plot(x, y1, '-r',label='Best Fit Line')
plt.scatter(x, y)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
def fit(x,y):
    x_num = x-x_mean
    y_num = y-y_mean
    m=(sum(x_num*y_num))/(sum(x_num*x_num))
    b=y_mean-(m*x_mean)
    return m,b
    
def predict(val):
    m,b=fit(x,y)
    return (m*val)+b
    
```

```python
fit(x,y)
```

```python
predict(7.8)
```
